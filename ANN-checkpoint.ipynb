```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
an=pd.read_csv("forestfires.csv")
an
```

```python
an.shape
```

```python
an.isna().sum()
```

```python
duplicate=an[an.duplicated()]
print("Duplicate Row:")
```

```python
an.columns
```

```python
an=an.loc[:,['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind','rain', 'area', 'dayfri', 'daymon', 'daysat', 'daysun', 'daythu','daytue', 'daywed', 'monthapr', 'monthaug', 'monthdec', 'monthfeb','monthjan', 'monthjul', 'monthjun', 'monthmar', 'monthmay', 'monthnov','monthoct', 'monthsep', 'size_category']]
an
```

```python
x=an.iloc[:,0:-1]
x
```

```python
y=an.iloc[:,-1]
y
```

```python
y.replace(to_replace=['small','large'],value=[0,1],inplace=True)
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=10)
```

# CREATING MODEL

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(15,input_dim=28,activation ='sigmoid'))
model.add(tf.keras.layers.Dense(10,activation='sigmoid'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x,y,validation_split=0.20, epochs=50, batch_size=100)
```

```python
model.save_weights('mywt.kmw')
```

```python
scores=model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('accuracy')
plt.ylabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('loss')
plt.ylabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
